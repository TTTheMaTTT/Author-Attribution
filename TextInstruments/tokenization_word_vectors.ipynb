```python
def tokenize(text, word2idx=None, padding_idx=0):
    tokenized = []
    for sentence in nltk.sent_tokenize(text):
        tokenized.append(nltk.word_tokenize(sentence))
    if word2idx is not None:
        tokenized = [[word2idx.get(word, 0) for word in sent] for sent in tokenized]
    return tokenized
```
